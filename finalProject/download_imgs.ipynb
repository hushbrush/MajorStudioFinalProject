```python
import pandas as pd
import requests
import os
import mimetypes
```

```python
data= pd.read_json("/Users/harshitachakravadhanula./Desktop/Parsons work/sem1/Code/MajorStudioFinalProject/finalProject/public/data.json", orient='records')
```

```python
data.head()
```

```python
# Extract the 'guid' from the first dictionary in the list in 'thumbnail'
data['thumbnail_link'] = data['thumbnail'].apply(lambda x: x[0]['guid'] if isinstance(x, list) and x else None)
```

```python
data['thumbnail_link'] 
```

```python

folder_path = "/Users/harshitachakravadhanula./Desktop/Parsons work/sem1/Major Studio 1/imgs"
os.makedirs(folder_path, exist_ok=True)

# Map content types to proper extensions
CONTENT_TYPE_TO_EXTENSION = {
    "image/jpeg": "jpg",
    "image/png": "png",
    "image/gif": "gif",
    "image/webp": "webp",
    "image/bmp": "bmp",
    "image/tiff": "tiff"
}

for index, row in data.iterrows():
    image_url = row['thumbnail_link']
    image_id = str(row['id'])

    try:
        response = requests.get(image_url, timeout=10)
        response.raise_for_status()

        content_type = response.headers.get('Content-Type', '')
        extension = CONTENT_TYPE_TO_EXTENSION.get(content_type.split(';')[0], None)

        if not extension:
            print(f"Unknown or unsupported content type for {image_url}: {content_type}")
            continue

        image_path = os.path.join(folder_path, f"{image_id}.{extension}")
        with open(image_path, 'wb') as f:
            f.write(response.content)

        print(f"Downloaded: {image_id}.{extension}")
    except Exception as e:
        print(f"Failed to download {image_url}: {e}")
```
